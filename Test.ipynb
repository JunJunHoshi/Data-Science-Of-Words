```python
import numpy as np
import pandas as pd
file=open('ForChild.txt','r',encoding='utf-8')
text=file.read()
text=text.replace('”',' ')
text=text.replace(',',' ')
text=text.replace('.',' ')
text=text.replace(';',' ')
text=text.replace("'"," ")
text=text.replace('"',' ')
text=text.replace('?',' ')
text=text.replace('!',' ')
words=[]
words=text.split()
print(len(words))
print(words)
```

```python
count=np.zeros(len(words))
wordslist=[]
for i in range(len(words)):
    words[i]=words[i].lower()
    count[i]=float((words.count(words[i]))/len(words)*1000)
    wordslist+=[(words[i],count[i])]
wordslist=set(wordslist)
x=[]
for i in wordslist:
    x+=[i]
print(len(x))
print(x)
```

```python
filea=open('ForAdult.txt','r',encoding='utf-8')
texta=filea.read()
texta=texta.replace('”',' ')
texta=texta.replace(',',' ')
texta=texta.replace('.',' ')
texta=texta.replace(';',' ')
texta=texta.replace("'"," ")
texta=texta.replace('"',' ')
texta=texta.replace('?',' ')
texta=texta.replace('!',' ')
wordsa=[]
wordsa=texta.split()
print(len(wordsa))
```

```python
counta=np.zeros(len(wordsa))
wordlista=[]
for i in range(len(wordsa)):
    wordsa[i]=wordsa[i].lower()
    counta[i]=float((wordsa.count(wordsa[i]))/len(wordsa)*1000)
    wordlista+=[(wordsa[i],counta[i])]
wordlista=set(wordlista)
y=[]
for j in wordlista:
    y+=[j]
print(len(y))
print(y)
```

```python
flugx=np.zeros(len(x))
flugy=np.zeros(len(y))
wordlist=[]

for i in range(len(x)):
    for j in range(len(y)):
        if x[i][0]==y[j][0]:
            wordlist+=[(x[i][0],x[i][1],y[j][1])]
            flugx[i]+=1
            flugy[j]+=1
            
for i in range(len(x)):
    if flugx[i]==0:
        wordlist+=[(x[i][0],x[i][1],0)]
for j in range(len(y)):
    if flugy[j]==0:
        wordlist+=[(y[j][0],0,y[j][1])]
print(wordlist)
```

```python
length=len(wordlist)
w=[]
x_child=np.zeros(length)
y_adult=np.zeros(length)
for i in range(length):
    w+=[(wordlist[i][0])]
    x_child[i]+=wordlist[i][1]
    y_adult[i]+=wordlist[i][2]

df=pd.DataFrame({
    'word' : w,
    'x(child)':x_child,
    'y(adult)':y_adult
})
df.head(10)
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
df_x=df.loc[:,['x(child)']]
df_y=df.loc[:,['y(adult)']]
```

```python
plt.scatter(df_x,df_y)
plt.title('The Relation Of Frequency')
plt.xlim(-5,40)
plt.ylim(-5,40)
plt.xlabel('The Number Of Words In Kids Books')
plt.ylabel('The Number Of Words In Adult Books')
```

```python
frequent_difference=[]
for i in wordlist:
    frequent_difference+=[(i[1]-i[2])**2/(i[1]+i[2])]
print(frequent_difference)
```

```python
df_fd=pd.DataFrame({
    'FrequencyDifference': frequent_difference
})
dif_df=pd.concat([df,df_fd],axis=1)
dif_df.sort_values(by=['FrequencyDifference'],ascending=False)
```

```python
realx=np.array(dif_df.loc[:,['x(child)']])
realy=np.array(dif_df.loc[:,['y(adult)']])
reald=np.array(dif_df.loc[:,['FrequencyDifference']])
for i in range(len(reald)):
    if reald[i]>1.2:
        plt.scatter(realx[i],realy[i],c='r')
    else:
        plt.scatter(realx[i],realy[i],c='b')
plt.title('The Relation Of Frequency')
plt.xlim(-5,40)
plt.ylim(-5,40)
plt.xlabel('The Frequency Of Words In Kids Books')
plt.ylabel('The Frequency Of Words In Adult Books')
x=np.linspace(-5,100,10)
y=x
plt.plot(x,y,c='y')
```
